```python
import csv
import matplotlib.pyplot as plt
```

```python
def getItems(source):
    with open(source, newline='', encoding='utf-8-sig') as csvfile:
        items = []

        reader = csv.reader(csvfile, delimiter=' ', skipinitialspace=True, quoting=csv.QUOTE_NONNUMERIC)
        for row in reader:
            items.append(row)
        return items
```

```python
clusteredData = getItems('logs/2b_best.csv')
for row in clusteredData:
    plt.scatter(x=row[0], y=row[1], c=row[2])

plt.scatter(x=-1.75, y=-1.0, c = (0.25, .84, .95))
plt.scatter(x=0.25, y=0.0, c = (0.25, .84, .95))
plt.scatter(x=1.50, y=1.0, c = (0.25, .84, .95))
plt.show()
```
